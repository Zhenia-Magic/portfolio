```r
library(pivottabler)
```

Now we need to create pivot table for all countries (1st level index), all missions (2nd level index), and all time periods (months) (3rd level index). Values in the pivot table will be the sum of female and male personnel for this cell.

### Important! 
The pivot table is quite complex so it can take up to 10 min to be created and rendered.

```r
pt <- PivotTable$new()
pt$addData(data_mission)
pt$addRowDataGroups("Contributing_Country",addTotal=FALSE)
pt$addRowDataGroups("Personnel_Type", addTotal=FALSE) 
pt$addRowDataGroups("Last_Reporting_Date", addTotal=FALSE) 
pt$defineCalculation(calculationName="TotalPersonnel", summariseExpression="sum(Female_Personnel + Male_Personnel)")
pt$renderPivot()
```

Then, we define a function fx that will get the value from the cell in the pivot table and apply it to all cells in the pivot table. After that, we can find different statistics for the number of personnel in MINUSMA mission.

```r
fx <- function(x) {x$rawValue} #create a function that will extract the value in the cell
numbers <- sapply(pt$getCells(columnNumbers=1), fx) #apply the function to all cells in the pivot table
mean(numbers)
median(numbers)
quantile(numbers)
```

```r
#create function to show the dates on which the specific number of personnel was positioned
# and highlight these dates in the pivot table
show_data <- function(pt, raw_val){
f_big = function(x) {if (x$rawValue == raw_val){x$rowNumber}} #create function to extract row number of the cell
index <- sapply(pt$getCells(columnNumbers=1), f_big) #apply the function to all cells
indexes <- c() 
#put the indexes of cells with the needed value to a vector
for (i in index){
    if (!is.null(i)){
        indexes <- c(indexes, i)
    }
} 
#extract the date from the pivot table
for (i in indexes){
grp <- pt$getLeafRowGroup(r=i)
print(grp$caption)}
cells <- pt$findCells(minValue=raw_val, maxValue=raw_val) #find cells with specific value
pt_sub <- pt #copy pivot table to not change the original one
pt_sub$setStyling(cells=cells, declarations=list("background-color"="cyan")) #highlight cells with specific value
pt_sub$renderPivot()} #show pivot table
```

```r
data_gdp <- read.csv("https://tinyurl.com/gdpcapitaourworldindata",na.strings=c("","NA"))
data_gdp 
```

```r
data_2017 <- data[data$Last_Reporting_Date >= "2017-12-01" & data$Last_Reporting_Date <= "2017-12-31",]
data_2017$overall_personnel <- data_2017$Female_Personnel + data_2017$Male_Personnel 
data_2017 <- aggregate(overall_personnel ~ Contributing_Country, data_2017, sum)
gdp_2017 <- data_gdp[data_gdp$Year == 2017,]
colnames(gdp_2017) <- c("Contributing_Country", 'Code', 'Year', 'GDP.per.capita')
k <- merge(data_2017, gdp_2017, by=c("Contributing_Country"))
k
```

```r
library(tidyverse)
data_gdp_personnel <- k %>% select(Contributing_Country, GDP.per.capita, overall_personnel)
data_gdp_personnel
```

```r
cor(data_gdp_personnel$GDP.per.capita, data_gdp_personnel$overall_personnel)
```

```r
linearMod <- lm(overall_personnel ~ GDP.per.capita, data=data_gdp_personnel)  # build linear regression model on full data
print(linearMod)
summary(linearMod)
```

```r
summary(linearMod)$coefficients[1, 1]
summary(linearMod)$coefficients[2, 1]
```

The equation obtained from linear regression is:

overall_personnel  -0.025 * GDP.per.capita + 1311.945

This equations shows that there is a small negative correlation between GDP per capita and number of personnel. Low p-values (less than 0.05) suggests that both point estimates (for b0 and b1) are stastically significant (so we can reject the null hypotheses that GDP does not affect the response or that the intercept is 0).

```r
plot(data_gdp_personnel$overall_personnel, data_gdp_personnel$GDP.per.capita, ylab = "GDP per capita", xlab = "Overall Personnel",pch = 16, cex = 1.3, col = alpha("pink", 0.8), main = "Linear Regression plot")
abline(lm(data_gdp_personnel$overall_personnel ~ data_gdp_personnel$GDP.per.capita), col = "blue")
```
